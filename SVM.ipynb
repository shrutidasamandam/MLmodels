```python
#importing the two libraries, numpy and the matplotlib for visualization
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
#make up data, which make_blobs does
from sklearn.datasets import make_blobs
# 40 lines of data, with 2 groups that have 20 different data items
X, y = make_blobs(n_samples = 40, centers = 2, random_state = 20)
# only 2 dimensions for the hyperplane to be linear
clf = svm.SVC(kernel = 'linear', C = 1)
# clf is created and then we can display it
clf.fit(X,y)
# make_blobs returns an numpy array
plt.scatter(X[:,0],X[:,1], c=y, s=30, cmap = plt.cm.Paired)
#plt.show()
```

```python
#predicting unknown data
newData = [[3,4],[5,6]]
print(clf.predict(newData))
# 0 means the [3,4] will be on one side and the [5,6] will be on the other side
```

```python
# plotting the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
# creating the grid to evaluate the model
xx = np.linspace(xlim[0],xlim[1],30)
yy = np.linspace(ylim[0],ylim[1],30)
YY, XX = np.meshgrid(yy,xx)
xy = np.vstack([XX.ravel(),YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# decision boundary & margins
ax.contour(XX,YY,Z,colors = 'k', levels = [-1, 0 ,1],
          alpha = 0.5,
          linestyles = ['--','-','--'])
#plot support vectyors
ax.scatter(clf.support_vectors_[:,0],
           clf.support_vectors_[:,1], s = 100,
           linewidth=1, facecolors = 'none')
plt.show()
```
