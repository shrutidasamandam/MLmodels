```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
np.random.seed(0)
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

```python
#generating a random number for each row, if it is less than .75 it is true
df['is_train'] = np.random.uniform(0,1,len(df))<=.75
df.head()
```

```python
# 2 new data frames, shows how many are trained vs test modules
train, test = df[df['is_train']==True], df[df['is_train']==False]
print('Number of observations in the training data:',len(train))
print('Number of observations in the test data:',len(test))
```

```python
#first 4 columns which are known as features
features = df.columns[:4]
print (features)
```

```python
# converting the species names into digits,represents the different type of flowers 
y = pd.factorize(train['species'])[0]
y
```

```python
# prioritize with n_jobs, random_state is how it starts
clf = RandomForestClassifier(n_jobs = 2, random_state = 0)
# training the set, the target is the Y 
clf.fit(train[features],y)
```

```python
clf.predict(test[features])
```

```python
#only looking at the first 10, generating a larger field
clf.predict_proba(test[features])[10:20]
```

```python
#mapping predictions to the names
# what the forest is actually doing
preds = iris.target_names[clf.predict(test[features])]
preds[0:5]
```

```python
# what the data is 
test['species'].head()
```

```python
pd.crosstab(test['species'],preds,rownames = ['Actual species'],
           colnames = ['Predicted Species'])
```
